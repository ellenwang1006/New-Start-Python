```python
# Input bags
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Input data
path='C:\\Users\\ellen\\Documents\\2019\\Python\\python入门笔记（强烈推荐）\\线性回归及非线性回归\线性回归以及非线性回归\\线性回归以及非线性回归\\Delivery.csv'
data=pd.read_csv(path,header=None)
x=data.iloc[:,0:2]
y=data.iloc[:,-1]
print(x)
print(y)
```

```python
model=LinearRegression()
model.fit(x,y)
```

```python
print('coefs are:',model.coef_)
print('intercept is:',model.intercept_)
x_test=[[102,4]]
predict=model.predict(x_test)
print('Predict is:', predict)
```

```python
# 3D plot

ax=plt.figure().add_subplot(111,projection='3d')
ax.scatter(x.iloc[:,0],x.iloc[:,1],y,c='r',marker='o',s=100)
x0=x.iloc[:,0]
x1=x.iloc[:,1]
x0,x1=np.meshgrid(x0,x1)
z=model.intercept_+ model.coef_[0]*x0 + model.coef_[1] * x1

ax.plot_surface(x0,x1,z)

ax.set_xlabel('Miles')
ax.set_ylabel('Times')
ax.set_zlabel('Duration')

plt.show()

```
