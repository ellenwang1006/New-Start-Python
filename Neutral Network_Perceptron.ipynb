```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Input Data
X= np.array([[1,3,3],
            [1,4,3],
            [1,1,1],
            [1,0,2]])

Y= np.array([[1],
             [1],
             [-1],
             [-1]])

# Set original coef
# np.random.random: Return random floats in the half-open interval [0.0, 1.0).

W= (np.random.random([3,1])-0.5)*2
print(W)

# learning rate
lr= 0.11
# Set output 
O=0

# Define updated coef
def update():
    global X,Y,W,lr
    O = np.sign(np.dot(X,W))
    W_C = lr * (X.T.dot(Y-O))/int(X.shape[0])
    W= W+ W_C
```

```python
for i in range(100):
    update()
    
    #[3,3],[4,3]
    x1 = [3,4]
    y1 = [3,3]

    #[1,1],[0,2]
    x2 = [1,0]
    y2 = [1,2]

    #Calculate  decision boundry
    k = -W[1]/W[2]
    d = -W[0]/W[2]

    # W0+W1*X+W2*Y=0
    # k= -W1/W2
    # b= -W0/W2

    print('k=',k)
    print('d=',d)

    xdata = (0,5)

    plt.figure()
    plt.plot(xdata,xdata*k+d,'r') 
    plt.scatter(x1,y1,c='b') #scatter(x,y)
    plt.scatter(x2,y2,c='y')
    plt.show()
```
