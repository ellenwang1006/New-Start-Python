```python
# Input bags
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Input data
path='C:\\Users\\ellen\\Documents\\2019\\Python\\python入门笔记（强烈推荐）\\线性回归及非线性回归\线性回归以及非线性回归\\线性回归以及非线性回归\\data.csv'
data=pd.read_csv(path,header=None,names=['x','y'])
x=data['x']
y=data['y']
plt.scatter(x,y)
plt.show()
```

```python
# theta 0
b=0
# theta 1
k=0
# learning rate
lr=0.0003
# iteration
epochs=100

# Cost Function
def compute_error(b,k,x,y):
    totalerror=0
    for i in range(0,len(x)):
        totalerror += ((k * x[i] + b)-y[i])**2
    return totalerror/float(len(x))/2.0
    
def gradient_descent_runner(x,y,b,k,lr,epochs):
    m=float(len(x))
    
    for i in range(epochs):
        b_grad=0
        k_grad=0
        for j in range(0,len(x)):
            b_grad +=(1/m) * ((k * x[j])+b-y[j])
            k_grad +=(1/m) * ((k * x[j])+b-y[j])* (x[j])
        b=b - lr * b_grad
        k=k - lr * k_grad
        
        #plot every 5 times
        if i % 5== 0:
            print("epochs:",i)
            plt.plot(x,y,'b.')
            plt.plot(x,k * x +b,'r.' )
            plt.show()
    return b,k
```

```python
print("Starting b={0},k={1},error={2}".format(b,k, compute_error(b, k, x, y)))
print('Running...')
b,k=gradient_descent_runner(x,y,b,k,lr,epochs)
print("After {0} times iteration,b={1},k={2},error={3}".format(epochs,b,k,compute_error(b,k,x,y)))
```
